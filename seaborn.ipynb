```python
import seaborn as sns

# display flights dataset as data frame. 
data = sns.load_dataset("flights")
data
```

Some datasets in seaborn
tips, iris, titanic, flights, diamonds, exercise and planets 

```python
# display specific data.

passengers_feb = data[data["month"]=="Feb" ]
passengers_feb
```

```python
# draw line chart 
import matplotlib.pyplot as plt
sns.set_style("darkgrid")
sns.lineplot(x="year", y="passengers", data=passengers_feb)
plt.title("Chart between Year and Passengers")
plt.ylabel("Passengers")
```

```python
sns.lineplot(x="year", y="passengers", data=data, size="month" , hue="month")
```

```python
diamonds = sns.load_dataset("diamonds")
diamonds

expensive_diamonds =diamonds[diamonds["price"]>18000]
expensive_diamonds


# plt.figure(figsize=(10,10))
sns.lineplot(x="cut", y="price", data=diamonds)
```

```python
fmri = sns.load_dataset("fmri")
fmri

```

```python
sns.relplot(data=fmri, x="timepoint", y="signal",hue="event", kind='line', style='region' , col='region', row="event")

```

```python
a = fmri[(fmri["event"]=="stim") & (fmri["region"] == "parietal")]
a
plt.title("relation between signal and time point when region = parietal event = stim ")
sns.lineplot(x="timepoint", y="signal", data=a)
```
