# load data from seaborn

```python
# load dataset about flights
data = sns.load_dataset("flights")
data
```

```python
new_data = data[data["year"] ==1950]
new_data

```

```python
# change size of figure 
sns.set_style("dark")
plt.figure(figsize=(10,10))
plt.title("Line between passengers and year")

# draw line plot using seaborn 
sns.lineplot(x="year", y="passengers", data=data, size="month", hue="month")
plt.ylabel("Passengers")

# change interval of axis.
plt.locator_params(axis='y', nbins=80) 
plt.locator_params(axis='x', nbins=10) 

```
